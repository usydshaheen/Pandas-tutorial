```python
# Create two sample DataFrames
data1 = {'A': [1, 2, 3, 4, 5], 'B': [6, 7, 8, 9, 10]}
data2 = {'A': [2, 4, 6, 8, 10], 'B': [1, 3, 5, 7, 9]}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
# Combine the two DataFrames into a list
dfs = [df1, df2]
```

```python
# Create a boxplot for each column in the list of DataFrames
fig, ax = plt.subplots()
ax.boxplot([df[col] for df in dfs for col in df.columns], labels=[col for df in dfs for col in df.columns])
ax.set_title('Boxplot of Multiple DataFrames')
ax.set_xlabel('Column')
ax.set_ylabel('Value')
```

```python
# Create a sample DataFrame with multiple groups
data = {'Group': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
        'Value': [1, 2, 3, 4, 5, 6, 7, 8, 9]}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Create a boxplot for each group
fig, ax = plt.subplots()
df.boxplot(by='Group', column='Value', ax=ax)
ax.set_title('Boxplot of Multiple Groups')
ax.set_xlabel('Group')
ax.set_ylabel('Value')

plt.show()
```
